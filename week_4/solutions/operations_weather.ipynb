```python
import numpy as np
```

```python
years = np.loadtxt('../assets/bristol_rainfall.csv', dtype='int', delimiter=',', usecols=0, skiprows=1)
rainfall = np.loadtxt('../assets/bristol_rainfall.csv', dtype='float', delimiter=',', usecols=1, skiprows=1)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(years,rainfall)

ax.title.set_text('Bristol historic rainfall data')
ax.set(xlabel='Year', ylabel='Annual rainfall [mm]')
```

```python
np.mean(rainfall)
```

```python
np.std(rainfall)
```

```python
def variation(x):
    return np.std(x) / np.mean(x)
```

```python
variation(rainfall)
```

```python
variation([50, 60, 51, 49, 53])
```
